### Cleaning the Entire dataset

```python
df.columns=['Country','Year','Status','LE','Adult Mortality','Infant Deaths','Alcohol','Percentage Expenditure','Hepatitis B','Measles','BMI','Under-Five Deaths','Polio','Total Expenditure','Diphtheria','HIV/AIDS','GDP','Population','Thinness 1_19_yrs','Thinness 5_9_yrs','Income composition of resources','Schooling']
#df[df.columns]
```

```python
# checking for missing values
df.isnull().sum()

#removing missing values
df.dropna(inplace=True)

#checking for duplicates
df.duplicated().sum()

#remove duplicates
df.drop_duplicates(inplace=True)

#checking for outlier
df.describe()

#remove outlier
df=df[(df['LE']>30)&(df['LE']<100)]
df.describe()
```

# Data Visualization

```python
# Distribution of Life Expectancy across different countries.

plt.figure(figsize=(16, 8))
sns.histplot(data=df, x='LE', hue='Status')
plt.xlabel('Life Expectancy')
plt.ylabel('Count')
plt.title('Distribution of Life Expectancy Across Different Countries')
plt.show()
```

###### From the histogram, we can see that the distribution of life expectancy is roughly normal, with a mean around 70 years and a range from around 45 to 90 years.

```python
# Differences in Life Expectancy by Region
# Investigating if there are any significant differences in life expectancy between different regions of the world

plt.figure(figsize=(16, 8))
sns.boxplot(data=df, x='Status', y='LE')
plt.xticks(rotation=90)
plt.xlabel('Development Status')
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy and Country Development Status')
plt.show()
```

###### From the histogram, we can see that the distribution of life expectancy is roughly normal, with a mean around 70 years and a range from around 45 to 90 years.

### Correlation between GDP per capital and Life Expectancy

```python
#Filter out the missing values
data = df.dropna(subset=['GDP','LE'])

#Extract the column of interest
gdp = data['GDP']
life_expectancy = data['LE']

# Scatterplot of GDP per capita vs Life Expectancy
plt.figure(figsize=(16,8))
sns.scatterplot(data=df, x="GDP", y="LE", hue='Status')
plt.title('Relationship between GDP per capital and Life Expectancy')
plt.xlabel('GDP per capital')
plt.ylabel('Life Expectancy')
plt.show()

#calculate the correlation coefficient
corr = df['GDP'].corr(df['LE']).round(3)
print(f"Correlation Coefficient: {corr}")
```

###### From the scatter plots, we can see that there is a positive correlation between GDP  and Life Expectancy. Countries with higher GDP tend to have higher life expectancies.

```python
# Economic and Social Factors vs Life Expectancy
# We can also investigate the relationship between economic and social factors and life expectancy by creating scatter plots.

plt.figure(figsize=(16, 8))

sns.scatterplot(data=df, x='Schooling', y='LE', hue='Status')
plt.xlabel('Schooling')
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy vs Schooling')
plt.show()
```

###### From the scatter plots, we can see that there is a positive correlation between Schooling  and Life Expectancy. Countries with higher high level of education tend to have higher life expectancies.

### Correlation between Alcohol and Life Expectancy

```python
plt.figure(figsize=(16,8))
sns.scatterplot(data=df, x="Alcohol", y="LE", hue='Status')
plt.title('Relationship between Alcohol and Life Expectancy')
plt.xlabel('A l c o h o l')
plt.ylabel('Life Expectancy')
plt.show()

#calculate the correlation coefficient
corr = df['Alcohol'].corr(df['LE']).round(3)
print(f"Correlation Coefficient: {corr}")
```

### Line Plot of Life Expectancy over Time in Egypt

```python
ghana = df[df['Country'] == 'Ghana']

#Create a line plot of life expectancy over time
plt.figure(figsize=(16,8))
plt.plot(ghana['Year'], ghana['LE'])
plt.title('Life Expectancy in Ghana (2000 - 2015)')
plt.xlabel('Y e a r')
plt.ylabel('Life Expectancy')
plt.show()
```

### Life Expectancy and GDP over Time in Ghana (2000 - 2015)

```python
gh = df[df['Country'] == 'Ghana']

color_tem = "#69b3a2"
color_pri = "#3399e6"

#Create a line grapy of life expectancy and GDP
fig,ax=plt.subplots(figsize=(16,8))
ax2 = ax.twinx()

fig.suptitle("Life Expectancy and GDP over Time in Ghana (2000 - 2015)", fontsize=14);
#fig.autofmt_xdata()

ax.plot(gh.Year, gh.LE, color=color_tem, lw=3)
ax.set_xlabel('Y e a r')
ax.set_ylabel('Life Expectancy', color=color_tem)
ax.tick_params(axis='y', labelcolor=color_tem)

ax2.plot(gh.Year, gh.Schooling, color=color_pri, lw=4)
ax2.set_ylabel('G D P', color=color_pri)
ax2.tick_params(axis='y', labelcolor=color_pri)
plt.show()
```

### Mean Life Expectancy by Year

```python
grouped_data = df.groupby(['Year'])[['LE', 'GDP']].mean()

#sort the data by mean GDP
sorted_data = grouped_data.sort_values('GDP', ascending=False)

#creating bar chart
plt.figure(figsize=(16,8))
plt.bar(sorted_data.index, sorted_data['LE'])
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Life Expectancy')
plt.title('Mean Life Expectancy by Year')
plt.show()
```
